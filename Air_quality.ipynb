```python
import numpy as np
import pandas as pd
```

```python
# loading the data from csv file to pandas dataframe
air_quality_data = pd.read_csv('AirQualityUCI.csv')
air_quality_data.head()
```

```python
# separating the data
air_quality_data = pd.read_csv('AirQualityUCI.csv' , sep = ';',decimal =',')
air_quality_data.head()
```

```python
# removing the last two column from the data set
air_quality_data = air_quality_data.iloc[:,:-2]
air_quality_data.head()
```

```python
air_quality_data.shape
```

```python
# 9356 represent the last data in the dataframe and remaining rows are just null value

air_quality_data = air_quality_data.head(9357)
air_quality_data.tail()
```

```python
# getting some info about the data

air_quality_data.info()
```

```python
# checking the number of missing value in dataframe
air_quality_data.isnull().sum()
```

```python
# this shows that there are no missing value in dataset. 
# But the actual missing values are tagged wih '-200' this is described in uci repository documentation
```

```python
# counting number of time '-200' appear in dataset

air_quality_data.isin([-200]).sum(axis =0)
```

```python
# handling the missing value
# by replacing the '-200' with Nan then calculating the mean and replace the nan with mean value

air_quality_data = air_quality_data.replace(to_replace = -200, value=np.NAN)
air_quality_data.isnull().sum()
```

```python
air_quality_data.mean(numeric_only=True)
```

```python
# replacing the missing value withe mean value
air_quality_data.fillna(air_quality_data.mean(numeric_only=True))
air_quality_data.tail()
```

```python
# again checking the missing value
air_quality_data.isnull().sum()
```

```python
# two tasks can be performed on this dataset
# 1. Forecasting
# 2. Regression
```

```python
# Forecasting with FB Prophet Algorithm
# converting the date from DD/MM/YYYY to YYYY-MM-DD

date_info = pd.to_datetime(air_quality_data['Date'], dayfirst= True)

print(date_info)
```

```python
time_info =air_quality_data["Time"]
```

```python
time_info = time_info.apply(lambda x: x.replace('.', ':'))
print(time_info)
```

```python
print(type(date_info))
print(type(time_info))
```

```python
# combining  date and time info to a pandas datafrmae
date_time = pd.concat([date_info,time_info],axis = 1)
print(date_time)
```

```python
# converting two column in one column
date_time['ds'] = date_time['Date'].astype(str) + ' '+ date_time['Time'].astype(str)
print(date_time)
```

```python
# we have to convert object data type into date time formate
timestamp = pd.DataFrame()
```

```python
timestamp['ds']= pd.to_datetime(date_time['ds'])
timestamp.head()
```

```python
timestamp['y'] = air_quality_data['RH']
timestamp.head()
```

```python
import plotly
from prophet import Prophet
```

```python
# training the prophet model
m = Prophet()
m.fit(timestamp)
```

```python
future = m.make_future_dataframe(periods=365, freq ='h')
future.tail()
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
fig1 = m.plot(forecast)
```

```python
fig2 = m.plot_components(forecast)
```
